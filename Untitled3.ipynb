```python
import pandas as pd

# Загружаем файл
df = pd.read_excel("ASARV.xlsx")

# Протягиваем значения в колонке 'N'
df['Подразделение'] = df['Подразделение'].ffill()

# Сохраняем результат
df.to_excel("ASARV_new.xlsx", index=False)
print ('done')



import pandas as pd

# Загружаем файл
df = pd.read_excel("your_file.xlsx")

# Удаляем строки, где значение в колонке M отсутствует (пусто / NaN / пробелы)
df['M'] = df['M'].replace(r'^\s*$', pd.NA, regex=True)
df = df.dropna(subset=['M'])

# Сохраняем результат
df.to_excel("your_file_filtered.xlsx", index=False)
```
